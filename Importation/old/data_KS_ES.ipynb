```python
#List of file to collect
mypath = "data/2006/"
file_names = [f for f in listdir(mypath) if isfile(join(mypath, f))]
len(file_names)
```

```python
# importer les ipc secondaire de la classe et de l'année en cours
expect_classes_A61B_2006 = []
with open(r'exemple données/expectation_space/2009_A61B_expectation_IPC_class.txt', 'r') as fp:
    for line in fp:
        x = line[:-1]
        expect_classes_A61B_2006.append(x)
```

```python
print(expect_classes_A61B_2006)
```

```python
# Création de vecteurs avec 

#Getting files in this XXXX year related to main classes and other classes
knowledge_space_A61B = []

expectation_space_A61B_2006 = []

#current_date = 20090101
#dateLate = []

for i in tqdm(range(len(file_names[:10000]))):
    patent_path = mypath + file_names[i]
    with open(patent_path) as f:
        d = json.load(f)
        f.close()
    
    class_mainIPC = d['main_ipcr_label']
    class_main = class_mainIPC[0:4]
    #We are collecting all documents related to the main class - we distinguish them later by date
    if class_main == 'A61B':
        knowledge_space_A61B.append(file_names[i])
    
    #For the expectations states - we have one for each year since the class are not similar
    #2014
    if class_mainIPC in expect_classes_A61B_2006:
        expectation_space_A61B_2006.append(file_names[i])
    #if int(d['date_published']) > current_date and d['decision'] != 'ACCEPTED' and d['decision'] != 'REJECTED':
    #    dateLate.append(file_names[i])
    #    print(i)
```

```python
#Saving knowledge space per year in csv
patent_number, titles, backgrounds, claims, summary, abstract, ipc, labels = [], [], [], [], [], [], [], []
current_date = 20090101
for i in tqdm(range(len(knowledge_space_A61B))):
    patent_path = mypath + knowledge_space_A61B[i]
    with open(patent_path) as f:
        d = json.load(f)
        f.close()
    #Not taking patents that are not published yet
    if int(d['date_published']) < current_date:
        
        #Creating the lists for the other information
        patent_number.append(d['application_number'])
        titles.append(d['title'])
        backgrounds.append(d['background'])
        claims.append(d['claims'])
        summary.append(d['summary'])
        abstract.append(d['abstract'])
        ipc.append(d['main_ipcr_label'])
        labels.append(d['decision'])
    else:
        #If the date is superior, we still take accepted or rejected into account
        if d['decision'] == 'ACCEPTED' or d['decision'] == 'REJECTED':
            #Creating the lists for the other information
            patent_number.append(d['application_number'])
            titles.append(d['title'])
            backgrounds.append(d['background'])
            claims.append(d['claims'])
            summary.append(d['summary'])
            abstract.append(d['abstract'])
            ipc.append(d['main_ipcr_label'])
            labels.append(d['decision'])
```

```python
df_known = pd.DataFrame({'application_number': patent_number, 'title': titles, 'abstract':abstract,
                    'claims':claims, 'background': backgrounds, 'summary':summary, 'ipc':ipc, 'label': labels})
print(len(df_known))
df_known
```

```python
df_known.to_csv('exemple données/knowledge_space/20092006_A61B_knowledge_space_raw.csv', index=False)
```

```python
#Saving expectations space per year in csv
patent_number, titles, backgrounds, claims, summary, abstract, ipc, labels = [], [], [], [], [], [], [], []
current_date = 20090101
for i in tqdm(range(len(expectation_space_A61B_2006))):
    patent_path = mypath + expectation_space_A61B_2006[i]
    with open(patent_path) as f:
        d = json.load(f)
        f.close()
    
    #Not taking patnts that are not published yet
    if int(d['date_published']) < current_date:
        
        #Creating the lists for the other information
        patent_number.append(d['application_number'])
        titles.append(d['title'])
        backgrounds.append(d['background'])
        claims.append(d['claims'])
        summary.append(d['summary'])
        abstract.append(d['abstract'])
        ipc.append(d['main_ipcr_label'])
        labels.append(d['decision'])
    else:
        #If the date is superior to 2016, we still take accepted or rejected into account ???
        if d['decision'] == 'ACCEPTED' or d['decision'] == 'REJECTED':
            #Creating the lists for the other information
            patent_number.append(d['application_number'])
            titles.append(d['title'])
            backgrounds.append(d['background'])
            claims.append(d['claims'])
            summary.append(d['summary'])
            abstract.append(d['abstract'])
            ipc.append(d['main_ipcr_label'])
            labels.append(d['decision'])
```
