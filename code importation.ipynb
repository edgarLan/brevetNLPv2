```python
base_path = 'C:/Users/edgar/OneDrive/Bureau/Ecole/HEC/A24/BrevetNLP/data/Comp/'

for year in range(2011, 2019):  # 2019 is excluded
    # Construct the filename
    filename = f'{base_path}{year}.tar.gz'
    
    # Open the tar.gz file
    with tarfile.open(filename, 'r:gz') as file:
        # Use the iterator to avoid loading all members at once
        for i, member in enumerate(file):
            if i > 100:  # Change this number for how many files you want to test with
                break
            file.extract(member, 'C:/Users/edgar/OneDrive/Bureau/Ecole/HEC/A24/BrevetNLP/data')
    
    print(f'Test extracted first 100 files from: {filename}')
```

```python

base_path = 'C:/Users/edgar/OneDrive/Bureau/Ecole/HEC/A24/BrevetNLP/data/Comp/'

for year in range(2005, 2019):  # 2019 is excluded
    # Construct the filename
    filename = f'{base_path}{year}.tar.gz'
    
    # Start time before extraction
    start_time = time.time()

    # Open the tar.gz file
    with tarfile.open(filename, 'r:gz') as file:
        # Extract all files
        file.extractall('C:/Users/edgar/OneDrive/Bureau/Ecole/HEC/A24/BrevetNLP/data')

    # Calculate the time taken for this iteration
    elapsed_time = time.time() - start_time

    print(f'Extracted: {filename} in {elapsed_time:.2f} seconds')
```
