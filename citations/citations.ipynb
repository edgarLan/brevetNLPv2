```python

# Open the ZIP file and read the TSV file directly
with zipfile.ZipFile(filename, 'r') as zip_ref:
    # List files in the zip to find the TSV file
    for file_info in zip_ref.infolist():
        if file_info.filename.endswith('.tsv'):
            # Read the TSV file in chunks to manage memory usage
            with zip_ref.open(file_info.filename) as tsv_file:
                # Initialize chunk size (adjust based on memory availability)
                chunk_size = 100000  # Number of rows per chunk
                total_rows = 0  # Initialize row counter

                # Use a flag to control header writing
                header_written = False

                # Read the file in chunks and export to TSV
                for chunk in pd.read_csv(tsv_file, sep='\t', on_bad_lines='skip', chunksize=chunk_size):
                    # Append each chunk to the output file
                    chunk.to_csv(output_file, sep='\t', mode='a', header=not header_written, index=False)
                    header_written = True  # After the first write, header should not be written again
                    
                    # Update the total row count
                    total_rows += len(chunk)

                    # Optional: print progress every certain number of rows
                    if total_rows % 1000000 == 0:  # Adjust based on preference
                        print(f"Processed {total_rows} rows so far.")

                print(f"Exported {total_rows} rows to {output_file}.")
```

```python
# Initialize row count
total_rows = 0
print_frequency = 1000000

with zipfile.ZipFile(filename, 'r') as zip_ref:
    # Iterate through the files in the ZIP
    for file_info in zip_ref.infolist():
        if file_info.filename.endswith('.tsv'):
            # Open the TSV file
            with zip_ref.open(file_info.filename) as tsv_file:
                # Count the total number of rows
                for line in tsv_file:
                    total_rows += 1  # Increment row count for each line
                    # Print updated total_rows
                    if total_rows % print_frequency == 0:
                        print(f"Total rows so far in {file_info.filename}: {total_rows}")


            print(f"Final total rows in {file_info.filename}: {total_rows}")
```
